# Jour 2 — Phase 1 : Audit qualité

Objectif :
Analyser le dataset avant toute transformation :

- dimensions
- types des variables
- valeurs manquantes
- répartition de la cible (Churn)

Pourquoi ?
Un bon preprocessing commence toujours par un diagnostic.

```python
def audit_qualite(df):
    """
    Affiche un rapport de santé du dataset :
    - forme (lignes, colonnes)
    - types des colonnes
    - % de valeurs manquantes
    - répartition de la cible Churn
    """

    print("=== FORMES DU DATASET ===")
    print(f"Lignes, colonnes : {df.shape}")

    print("\n=== TYPES DES COLONNES ===")
    print(df.dtypes)

    # Valeurs manquantes
    print("\n=== VALEURS MANQUANTES (%) ===")
    manquants = df.isna().mean() * 100
    manquants = manquants.sort_values(ascending=False)

    # Afficher seulement celles > 0
    print(manquants[manquants > 0].round(2))

    # Répartition de la cible
    print("\n=== RÉPARTITION DE LA CIBLE (Churn) ===")

    if "Churn" in df.columns:
        counts = df["Churn"].value_counts()
        total = len(df)

        for valeur, count in counts.items():
            pourcentage = (count / total) * 100
            print(f"{valeur} : {count} ({pourcentage:.2f}%)")
    else:
        print(" Colonne Churn introuvable")
```

```python
audit_qualite(df)
```

```python
print("=== TEST 1 : cas normal ===")
audit_qualite(df)
```

```python
df_one_class = df[df["Churn"] == "No"]

print("\n=== TEST 2 : une seule classe ===")
audit_qualite(df_one_class)
```

```python
print("\n=== TEST 3 : vérification déséquilibre ===")
audit_qualite(df)
```

## Détection des outliers avec la règle IQR
``

```python
def detecter_outliers_iqr(df, colonne):
    Q1 = df[colonne].quantile(0.25)
    Q3 = df[colonne].quantile(0.75)
    IQR = Q3 - Q1

    borne_basse = Q1 - 1.5 * IQR
    borne_haute = Q3 + 1.5 * IQR

    outliers = df[(df[colonne] < borne_basse) | (df[colonne] > borne_haute)]

    print(f"\n--- {colonne} ---")
    print(f"Bornes : [{borne_basse:.2f}, {borne_haute:.2f}]")
    print(f"Nombre d'outliers : {len(outliers)}")
```

```python
for col in ["tenure", "MonthlyCharges", "TotalCharges"]:
    detecter_outliers_iqr(df, col)
```

```python
def cap_outliers(df, colonne):
    Q1 = df[colonne].quantile(0.25)
    Q3 = df[colonne].quantile(0.75)
    IQR = Q3 - Q1

    borne_basse = Q1 - 1.5 * IQR
    borne_haute = Q3 + 1.5 * IQR

    df[colonne] = df[colonne].clip(borne_basse, borne_haute)

    return df
```

## Analyse

Des valeurs aberrantes ont été détectées dans plusieurs colonnes, notamment :

- TotalCharges
- MonthlyCharges

Cependant :

- ces valeurs élevées ne sont pas forcément des erreurs
- elles peuvent correspondre à de vrais clients (factures élevées)

Choix effectué :
- conservation des outliers

Justification :
- supprimer ces valeurs ferait perdre de l'information importante
- dans un contexte métier, les clients avec de fortes charges sont pertinents

Conclusion :
les outliers sont conservés.


# Phase 5 — Corrélations et multicolinéarité

Objectif :
- analyser les relations entre variables
- détecter les variables redondantes
- réduire la multicolinéarité

Méthodes :
- heatmap de corrélation
- calcul du VIF

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Matrice de corrélation
corr_matrix = df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)

plt.title("Matrice de corrélation")
plt.show()
```

```python
# Trouver corrélations > 0.8
threshold = 0.8

high_corr = []

for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            corr = abs(corr_matrix.loc[col1, col2])
            if corr > threshold:
                high_corr.append((col1, col2, corr))

# Supprimer doublons
high_corr_unique = list(set([tuple(sorted([a, b]) + [c]) for a, b, c in high_corr]))

for item in high_corr_unique:
    print(item)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# On prend quelques colonnes importantes
cols = ["tenure", "MonthlyCharges", "TotalCharges"]

X = df[cols]

vif = pd.DataFrame()
vif["variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

```python
df = df.drop(columns=["TotalCharges"])
```

```python
cols = ["tenure", "MonthlyCharges"]

X = df[cols]

vif = pd.DataFrame()
vif["variable"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif)
```

## Analyse

Certaines variables étaient fortement corrélées, notamment :

- tenure
- MonthlyCharges
- TotalCharges

Cela crée un problème de multicolinéarité.

Observation :
- TotalCharges est fortement dépendante des deux autres variables

Choix effectué :
- suppression de TotalCharges

Justification :
- information redondante
- amélioration de la stabilité des modèles

Conclusion :
la multicolinéarité a été réduite.

# Phase 6 — Variables les plus discriminantes

Objectif :
Identifier les variables les plus importantes pour prédire le churn.

Méthodes utilisées :
- corrélation avec la cible
- importance via Random Forest

But :
comprendre les facteurs clés du churn.

```python
# Corrélation avec la cible
corr_target = df.corr()["Churn"].abs().sort_values(ascending=False)

print("Top 10 corrélations avec Churn :")
print(corr_target.head(10))
```

```python
from sklearn.ensemble import RandomForestClassifier

# Séparer X / y
X = df.drop(columns=["Churn"])
y = df["Churn"]

# Modèle
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Importance
importances = pd.Series(model.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=False)

print("Top 10 features (Random Forest) :")
print(importances.head(10))
```

## Comparaison des méthodes

On compare :
- corrélation directe
- importance via Random Forest

```python
top_corr = corr_target.head(10)
top_rf = importances.head(10)

print("=== Corrélation ===")
print(top_corr)

print("\n=== Random Forest ===")
print(top_rf)
```
